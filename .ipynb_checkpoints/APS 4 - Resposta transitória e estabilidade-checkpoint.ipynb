$$\frac{d\omega(t)}{dt} + 2 \cdot \omega(t) = 10 e_{a}(t)$$

Aplicando laplace

$$s\cdot W(s) + 2W(s) = 10 E_{a}(s)$$

$$W(s)\cdot(s+2) = 10 E_{a}(s)$$

$$\frac{W(s)}{E_{a}(s)} = \frac{10}{s+2}$$

Como é um sistema de 1ª ordem, pode-se comparar com o formato genérico:

$$G(s) = k\frac{a}{s+a}$$

Portanto:

$$\frac{W(s)}{E_{a}(s)} = 5\cdot \frac{2}{s+2}$$

- Tau

$$\tau = \frac{1}{a} \to 0.5s$$

- Tempo de acomodação

$$T = \frac{4}{a} = 2s $$

2.

a)

$$2\frac{d^2x(t)}{dt^2} + 20\frac{dx(t)}{dt} + 200x(t) = 100f(t)$$

Aplica-se Laplace

$$2s^2X(s) + 20sX(s) + 200X(s) = 100F(s)$$

$$X(s)\cdot (2s^2 + 20s + 200) = 100F(s)$$

$$\frac{X(s)}{F(s)} = \frac{100}{2s^2 + 20s + 200}$$

Froma genérica:

$$G(s) = k\frac{\omega_n^2}{s^2 + 2\zeta\omega_n + \omega_n^2}$$

$$G(s) = \frac{1}{2} \frac{100}{s^2 + 10s + 100}$$

$$\omega_n = \sqrt{100} \to 10$$

$$\zeta = \frac{10}{\omega_n*2} \to 0.5$$

$$k = 0.5$$

- Instante de pico:

$$T_P= \frac{\pi}{w_n\sqrt{1-\zeta^2}}\to 0.36275$$

- UP:

$$UP = e^{-\frac{\zeta\cdot \pi}{\sqrt{1-\zeta^2}}}\to 0.16303$$

- Ts:

$$T_s = \frac{4}{2\zeta\omega_n} \to 0.8s$$

![grafic](img/aps4_2b.png)

```python
tp = math.pi/(10*math.sqrt(1-0.5*0.5))
print(f"tp = 0.36275")
up = math.exp(-0.5*math.pi/math.sqrt(1-0.5*0.5))
print("up = 16.303% ")
ts = 4/(10*0.5)
print("ts = 0.8")
```

b)

$$\frac{d^2x(t)}{dt^2} + 6\frac{dx(t)}{dt} + 16x(t) = 64f(t)$$

Aplica-se Laplace:

$$s^2X(s) + 6sX(s) + 16X(s) = 64F(s)$$

$$X(s)\cdot(s^2+6s+16) = 64F(s)$$

$$\frac{X(s)}{F(s)} = \frac{64}{s^2 + 6s + 16}$$

Froma genérica:

$$G(s) = k\frac{\omega_n^2}{s^2 + 2\zeta\omega_n + \omega_n^2}$$

$$G(s) = 4\frac{16}{s^2 + 6s + 16}$$

$$\omega_n = \sqrt{16} \to 4$$

$$\zeta = \frac{6}{\omega_n*2} \to 0.75$$

$$k = 4$$

- Instante de pico:

$$T_P= \frac{\pi}{w_n\sqrt{1-\zeta^2}}\to 1.18741$$

- UP:

$$UP = e^{-\frac{\zeta\cdot \pi}{\sqrt{1-\zeta^2}}}\to 2.837%$$

- Ts:

$$T_s = \frac{4}{2\zeta\omega_n} \to 1.333s$$


![grafic](img/aps4_2a.png)

```python
tp = math.pi/(4*math.sqrt(1-0.75*0.75))
print(f"tp = {tp}")
up = math.exp(-0.75*math.pi/math.sqrt(1-0.75*0.75))
print(f"up = {up} ")
ts = 4/(4*0.75)
print(f"ts = {ts}")
```
